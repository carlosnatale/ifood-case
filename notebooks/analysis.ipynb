```sql
-- Interpretacao A: media por corrida, mes a mes
-- 'Quanto vale em media uma corrida em cada mes?'
SELECT
    ano_mes,
    ROUND(AVG(total_amount), 2) AS media_total_amount_por_corrida
FROM ifood_case.silver.yellow_trips
GROUP BY ano_mes
ORDER BY ano_mes;
```

```sql
-- Interpretacao B: media da receita mensal no periodo
-- 'Quanto a frota fatura em media por mes?'
WITH receita_por_mes AS (
    SELECT ano_mes, SUM(total_amount) AS receita_mes
    FROM ifood_case.silver.yellow_trips
    GROUP BY ano_mes
)
SELECT ROUND(AVG(receita_mes), 2) AS media_receita_mensal
FROM receita_por_mes;
```

```sql
SELECT
    hora_pickup                    AS hora_do_dia,
    ROUND(AVG(passenger_count), 2) AS media_passageiros,
    COUNT(*)                       AS total_corridas
FROM ifood_case.silver.yellow_trips
WHERE ano_mes = '2023-05'
GROUP BY hora_pickup
ORDER BY hora_pickup;
```
