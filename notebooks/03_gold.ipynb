```sql
CREATE OR REPLACE TABLE ifood_case.gold.receita_mensal
USING DELTA AS
SELECT
    ano_mes,
    COUNT(*)                    AS total_corridas,
    ROUND(SUM(total_amount), 2) AS receita_total,
    ROUND(AVG(total_amount), 2) AS ticket_medio_corrida
FROM ifood_case.silver.yellow_trips
GROUP BY ano_mes;
```

```sql
CREATE OR REPLACE TABLE ifood_case.gold.passageiros_por_hora
USING DELTA AS
SELECT
    ano_mes,
    hora_pickup,
    COUNT(*)                       AS total_corridas,
    ROUND(AVG(passenger_count), 2) AS media_passageiros
FROM ifood_case.silver.yellow_trips
GROUP BY ano_mes, hora_pickup;
```

```sql
-- Receita por mes
SELECT * FROM ifood_case.gold.receita_mensal ORDER BY ano_mes;
 
-- Passageiros por hora (maio)
SELECT * FROM ifood_case.gold.passageiros_por_hora
WHERE ano_mes = '2023-05'
ORDER BY hora_pickup;
```

```sql

-- Quantas versoes a tabela Silver tem?
DESCRIBE HISTORY ifood_case.silver.yellow_trips;

-- Como eram os dados na versao anterior (antes do ultimo overwrite)?
SELECT COUNT(*) FROM ifood_case.silver.yellow_trips VERSION AS OF 0;

-- Comparacao entre versoes: houve mudanca no volume?
SELECT 'versao_atual' AS versao, COUNT(*) AS linhas
FROM ifood_case.silver.yellow_trips
UNION ALL
SELECT 'versao_0', COUNT(*)
FROM ifood_case.silver.yellow_trips VERSION AS OF 0;
```
